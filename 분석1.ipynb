```python
# 대한민국 시 목록
city_list = [
    "서울특별시", "부산광역시", "대구광역시", "인천광역시", "광주광역시", "대전광역시", "울산광역시", "세종특별자치시",
    "수원시", "성남시", "의정부시", "안양시", "부천시", "광명시", "평택시", "동두천시", "안산시", "고양시", "과천시", 
    "구리시", "남양주시", "오산시", "시흥시", "군포시", "의왕시", "하남시", "용인시", "파주시", "이천시", "안성시", 
    "김포시", "화성시", "광주시", "양주시", "포천시", "여주시",
    "춘천시", "원주시", "강릉시", "동해시", "태백시", "속초시", "삼척시",
    "충주시", "제천시",
    "청주시", "충주시", 
    "천안시", "공주시", "보령시", "아산시", "서산시", "논산시", "계룡시", "당진시",
    "전주시", "군산시", "익산시", "정읍시", "남원시", "김제시",
    "목포시", "여수시", "순천시", "나주시", "광양시",
    "포항시", "경주시", "김천시", "안동시", "구미시", "영주시", "영천시", "상주시", "문경시", "경산시",
    "진주시", "통영시", "사천시", "김해시", "밀양시", "거제시", "양산시",
    "서귀포시", "제주시"
]

# 시 이름 추출 함수 정의
def extract_city_name(full_name):
    for city in city_list:
        if city in full_name:
            return city
    return None

# 시 이름 추출하여 새로운 열 추가
safety_data['시'] = safety_data['행정구역'].apply(extract_city_name)

# '시' 열이 None이 아닌 행만 필터링
safety_data = safety_data[safety_data['시'].notna()]

# 데이터 타입 확인 및 변환
safety_data[['교통사고', '화재', '범죄', '생활안전', '자살', '감염병']] = safety_data[['교통사고', '화재', '범죄', '생활안전', '자살', '감염병']].apply(pd.to_numeric, errors='coerce')

# 시별로 평균 등급 계산
grouped_safety_data_by_city = safety_data.groupby('시').mean(numeric_only=True).reset_index()

# 등급을 평균값으로 계산하여 소수점 첫째 자리까지 표시
grouped_safety_data_by_city = grouped_safety_data_by_city.round(1)

# 결과 데이터프레임 출력
grouped_safety_data_by_city
```

### Q1. 진주와 비슷한 경제적 요건을 가진 도시들에서 범죄등급이 낮은 도시는?

```python
# 진주와 경제적 요건이 비슷한 도시 목록
similar_cities = ["청주시", "천안시", "구미시", "익산시", "여수시", "김해시", "아산시", "양산시", "제천시", "경산시"]


# 시 이름 추출 함수 정의
def extract_city_name(full_name):
    parts = full_name.split()
    if len(parts) > 1:
        return parts[1]
    return parts[0]

# 시 이름 추출하여 새로운 열 추가
safety_data['시'] = safety_data['행정구역'].apply(extract_city_name)

# '시' 열이 None이 아닌 행만 필터링
safety_data = safety_data[safety_data['시'].notna()]

# 경제적 요건이 비슷한 시들 필터링
similar_cities_data = safety_data[safety_data['시'].isin(similar_cities)]

# 데이터 타입 확인 및 변환
similar_cities_data[['교통사고', '화재', '범죄', '생활안전', '자살', '감염병']] = similar_cities_data[['교통사고', '화재', '범죄', '생활안전', '자살', '감염병']].apply(pd.to_numeric, errors='coerce')

# 시별로 평균 등급 계산
grouped_similar_cities_data = similar_cities_data.groupby('시').mean(numeric_only=True).reset_index()

# 등급을 평균값으로 계산하여 소수점 첫째 자리까지 표시
grouped_similar_cities_data = grouped_similar_cities_data.round(1)

# 범죄 등급을 기준으로 정렬
sorted_similar_cities_data = grouped_similar_cities_data.sort_values(by='범죄')

# 결과 데이터프레임 출력
sorted_similar_cities_data
```
