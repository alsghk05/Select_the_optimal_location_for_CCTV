```python
import pandas as pd

# 'cp949' 인코딩을 사용하여 파일을 읽기
crime_place = pd.read_csv('cctv설치 현황, 범죄 발생 건수\(12-21)년도별 진주시 범죄 발생 상세 주소 + 동 주소 + 범죄 항목\경찰청_범죄 발생 장소별 통계_20211231.csv', encoding='cp949')
```

```python
crime_place
```

결측치 확인 및 처리

```python
pd.DataFrame(crime_place.isnull().sum())
```

데이터 타입 변환

```python
# 모든 값이 숫자형이어야 하므로, 숫자형으로 변환합니다.
crime_place.iloc[:, 2:] = crime_place.iloc[:, 2:].apply(pd.to_numeric)
```

```python
crime_place
```

불필요한 열 제거

필요한 열 생성

```python
# 예를 들어, 각 장소별 범죄 발생 총합을 계산하여 새로운 열을 생성합니다.
crime_place['총합'] = crime_place.iloc[:, 2:].sum(axis=1)
```

```python
crime_place
```

데이터 정렬 및 인덱싱

```python
# '총합' 기준으로 정렬
crime_place = crime_place.sort_values(by='총합', ascending=False)
```
